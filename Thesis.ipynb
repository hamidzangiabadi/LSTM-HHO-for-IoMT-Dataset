# OUR Decode Function 

```python
lb = np.array([16, 0.1, 1e-5])
ub = np.array([256, 0.5, 1e-2])

FIXED_BATCH_SIZE = 128
```

```python
import numpy as np

def decode_position(position):

    units = int(np.round(position[0]))
    dropout = float(position[1])
    learning_rate = float(position[2])
    
    units = int(np.clip(units, 16, 256))
    
    dropout = np.clip(dropout, 0.1, 0.5)
    learning_rate = np.clip(learning_rate, 1e-5, 1e-2)

    return {
        "units": units,
        "dropout": dropout,
        "learning_rate": learning_rate,
        "batch_size": FIXED_BATCH_SIZE
    }

```

# Creating the model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, LSTM, Dropout, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping

```

```python

def build_lstm_model(input_shape, units, dropout, learning_rate):
    model = Sequential()

    model.add(Input(shape=input_shape))

    model.add(
        LSTM(
            units=units,
            return_sequences=False,
            activation="tanh"
        )
    )

    model.add(Dropout(dropout))

    model.add(Dense(1, activation="sigmoid"))

    optimizer = Adam(learning_rate=learning_rate)

    model.compile(
        optimizer=optimizer,
        loss="binary_crossentropy",
        metrics=[tf.keras.metrics.Precision(), tf.keras.metrics.Recall()]
    )

    return model
```

# Fitness Function 

```python
from sklearn.metrics import f1_score, recall_score, precision_score
import numpy as np
import tensorflow as tf
import gc
```

```python
fitness_cache = {}

def fitness_function(position, class_weight_dict):

    key = tuple(np.round(position, 4))
    if key in fitness_cache:
        return fitness_cache[key]

    tf.random.set_seed(42)
    np.random.seed(42)

    params = decode_position(position)

    model = build_lstm_model(
        input_shape=(1, 45),
        units=params["units"],
        dropout=params["dropout"],
        learning_rate=params["learning_rate"]
    )

    es = EarlyStopping(
        monitor="val_loss",
        patience=1,
        restore_best_weights=True,
        verbose=0
    )

    model.fit(
        X_train_fit,
        y_train_fit,
        validation_data=(X_val_fit, y_val_fit),
        epochs=1,                
        batch_size=128,
        class_weight=class_weight_dict,
        callbacks=[es],
        verbose=0
    )

    y_prob = model.predict(X_val_fit, verbose=0).ravel()
    y_pred = (y_prob > 0.5).astype(int)

    rec = recall_score(y_val_fit, y_pred, zero_division=0)
    pre = precision_score(y_val_fit, y_pred, zero_division=0)

    fitness = 0.7 * rec + 0.3 * pre

    fitness_cache[key] = fitness

    tf.keras.backend.clear_session()
    del model

    return fitness
```

# Levy Flight

```python
import numpy as np
from scipy.special import gamma

def levy_flight(dim, beta=1.5):
    sigma_u = (
        gamma(1 + beta) * np.sin(np.pi * beta / 2) /
        (gamma((1 + beta) / 2) * beta * 2 ** ((beta - 1) / 2))
    ) ** (1 / beta)

    u = np.random.normal(0, sigma_u, dim)
    v = np.random.normal(0, 1, dim)

    step = u / (np.abs(v) ** (1 / beta))
    return step
```

# MHHO

```python
class MHHO:
    def __init__(self, fitness_function, lb, ub, dim, n_hawks=10, max_iter=10,
                 w_max=0.95, w_min=0.4, levy_alpha=0.15, record_curve=True):
        self.fitness_function = fitness_function
        self.lb = lb
        self.ub = ub
        self.dim = dim
        self.n_hawks = n_hawks
        self.max_iter = max_iter
        self.w_max = w_max
        self.w_min = w_min
        self.levy_alpha = levy_alpha
        self.record_curve = record_curve

        self.hawks = np.random.uniform(lb, ub, (n_hawks, dim))
        self.fitness = np.array([fitness_function(h) for h in self.hawks])

        idx = np.argmax(self.fitness)
        self.rabbit_pos = self.hawks[idx].copy()
        self.rabbit_fit = self.fitness[idx]

    def optimize(self):
        best_curve = [] if self.record_curve else None

        for t in range(self.max_iter):            
            omega = self.w_max - (self.w_max - self.w_min) * (t / self.max_iter)
            omega *= (0.9 + 0.2 * np.random.rand())
            omega = np.clip(omega, self.w_min, self.w_max)
            for i in range(self.n_hawks):
                E = (2 * np.random.rand() - 1) * (1 - t / self.max_iter)
                J = 2 * (1 - np.random.rand())
                X = self.hawks[i]

                if abs(E) >= 1:
                    rand_idx = np.random.randint(self.n_hawks)
                    X_rand = self.hawks[rand_idx]
                    X_hho = X_rand - np.random.rand() * np.abs(X_rand - 2 * np.random.rand() * X)

                else:
                    X_hho = self.rabbit_pos - E * np.abs(J * self.rabbit_pos - X)
                    
                    if 0.3 < abs(E) < 0.8 and np.random.rand() < 0.5:
                        levy = levy_flight(self.dim) 
                        X_hho = X_hho +  0.15 * levy

                X_new = omega * X + (1 - omega) * X_hho
                X_new = np.clip(X_new, self.lb, self.ub)

                fit_new = self.fitness_function(X_new)


                if fit_new > self.fitness[i]:
                    self.hawks[i] = X_new
                    self.fitness[i] = fit_new

            idx = np.argmax(self.fitness)
            if self.fitness[idx] > self.rabbit_fit:
                self.rabbit_fit = self.fitness[idx]
                self.rabbit_pos = self.hawks[idx].copy()

            if self.record_curve:
                best_curve.append(self.rabbit_fit)

            print(f"Iter {t+1}/{self.max_iter} | Best F1: {self.rabbit_fit:.5f}")

        curve_to_return = np.array(best_curve) if self.record_curve else np.array([])
        
        return self.rabbit_pos, self.rabbit_fit, curve_to_return
```

```python
def evaluate_mhho(
    fitness_function,
    lb,
    ub,
    dim,
    n_runs=5,
    n_hawks=10,
    max_iter=10
):
    best_scores = []

    for run in range(n_runs):
        print(f"\n========== Run {run+1}/{n_runs} ==========")

        mhho = MHHO(
            fitness_function=fitness_function,
            lb=lb,
            ub=ub,
            dim=dim,
            n_hawks=n_hawks,
            max_iter=max_iter
        )

        best_pos, best_fit = mhho.optimize()
        best_scores.append(best_fit)

        print(f"Run {run+1} Best F1: {best_fit:.5f}")

    best_scores = np.array(best_scores)

    print("\n========== MHHO Final Report ==========")
    print(f"Best  F1 : {best_scores.max():.5f}")
    print(f"Mean  F1 : {best_scores.mean():.5f}")
    print(f"Std   F1 : {best_scores.std():.5f}")

    return best_scores
```

# pure HHO

```python
class HHO:
    def __init__(self, fitness_function, lb, ub, dim,
                 n_hawks=10, max_iter=10, record_curve=False):

        self.fitness_function = fitness_function
        self.lb = lb
        self.ub = ub
        self.dim = dim
        self.n_hawks = n_hawks
        self.max_iter = max_iter
        self.record_curve = record_curve

        self.hawks = np.random.uniform(lb, ub, (n_hawks, dim))
        self.fitness = np.array([self.fitness_function(h) for h in self.hawks])

        idx = np.argmax(self.fitness)
        self.rabbit_pos = self.hawks[idx].copy()
        self.rabbit_fit = self.fitness[idx]

    def optimize(self):
        best_curve = []

        for t in range(self.max_iter):
            E = (2 * np.random.rand() - 1) * (1 - t / self.max_iter)

            for i in range(self.n_hawks):
                X = self.hawks[i]
                J = 2 * (1 - np.random.rand())

                if abs(E) >= 1:
                    rand_idx = np.random.randint(self.n_hawks)
                    X_rand = self.hawks[rand_idx]
                    X_new = X_rand - np.random.rand() * np.abs(X_rand - 2 * np.random.rand() * X)
                else:
                    X_new = self.rabbit_pos - E * np.abs(J * self.rabbit_pos - X)

                X_new = np.clip(X_new, self.lb, self.ub)
                fit_new = self.fitness_function(X_new)

                if fit_new > self.fitness[i]:
                    self.hawks[i] = X_new
                    self.fitness[i] = fit_new

            idx = np.argmax(self.fitness)
            if self.fitness[idx] > self.rabbit_fit:
                self.rabbit_fit = self.fitness[idx]
                self.rabbit_pos = self.hawks[idx].copy()

            best_curve.append(self.rabbit_fit)

            print(f"[HHO] Iter {t+1}/{self.max_iter} | Best F1: {self.rabbit_fit:.5f}")

        return self.rabbit_pos, self.rabbit_fit, np.array(best_curve)
```

# Run Tests

```python
import numpy as np
import random
import tensorflow as tf
import pandas as pd
from datetime import datetime
import time
import os

def set_seed(seed):
    np.random.seed(seed)
    random.seed(seed)
    tf.random.set_seed(seed)


N_RUNS = 5
dim = 3
N_HAWKS = 5
MAX_ITER = 30

timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")

DETAILED_FILE = f'detailed_results_{timestamp}.csv'
SUMMARY_RUN_FILE = f'summary_run_by_run_{timestamp}.csv'
STAT_FILE = f'statistical_comparison_{timestamp}.csv'

results_baseline, results_hho, results_mhho = [], [], []
execution_times = {'baseline': [], 'hho': [], 'mhho': []}
curves_hho, curves_mhho = [], []


for run in range(N_RUNS):

    seed = 100 + run
    set_seed(seed)

    print(f"\n================ RUN {run+1}/{N_RUNS} | Seed = {seed} ================")

    print("Running Baseline LSTM ...")
    start = time.time()
    f1_baseline = run_baseline_lstm()
    t_baseline = time.time() - start

    results_baseline.append(f1_baseline)
    execution_times['baseline'].append(t_baseline)

    print("Running HHO-LSTM ...")
    start = time.time()
    hho = HHO(
        fitness_function=lambda pos: fitness_function(pos, class_weight_dict),
        lb=lb, ub=ub, dim=dim,
        n_hawks=N_HAWKS, max_iter=MAX_ITER,
        record_curve=True
    )
    _, f1_hho, curve_hho = hho.optimize()
    t_hho = time.time() - start

    results_hho.append(f1_hho)
    execution_times['hho'].append(t_hho)
    curves_hho.append(curve_hho)

    np.save(f'curve_hho_run{run+1}.npy', np.array(curve_hho))

    print("Running MHHO-LSTM ...")
    start = time.time()
    mhho = MHHO(
        fitness_function=lambda pos: fitness_function(pos, class_weight_dict),
        lb=lb, ub=ub, dim=dim,
        n_hawks=N_HAWKS, max_iter=MAX_ITER,
        record_curve=True
    )
    _, f1_mhho, curve_mhho = mhho.optimize()
    t_mhho = time.time() - start

    results_mhho.append(f1_mhho)
    execution_times['mhho'].append(t_mhho)
    curves_mhho.append(curve_mhho)

    np.save(f'curve_mhho_run{run+1}.npy', np.array(curve_mhho))

    imp_hho = ((f1_hho - f1_baseline) / f1_baseline) * 100
    imp_mhho = ((f1_mhho - f1_baseline) / f1_baseline) * 100

    run_row = pd.DataFrame([{
        'Run': run + 1,
        'Seed': seed,
        'Baseline_F1': f1_baseline,
        'HHO_F1': f1_hho,
        'MHHO_F1': f1_mhho,
        'Baseline_Time_sec': t_baseline,
        'HHO_Time_sec': t_hho,
        'MHHO_Time_sec': t_mhho,
        'HHO_Improvement_%': imp_hho,
        'MHHO_Improvement_%': imp_mhho
    }])

    run_row.to_csv(
        DETAILED_FILE,
        mode='a',
        index=False,
        header=not os.path.exists(DETAILED_FILE),
        float_format='%.6f'
    )

    run_row.to_csv(
        SUMMARY_RUN_FILE,
        mode='a',
        index=False,
        header=not os.path.exists(SUMMARY_RUN_FILE),
        float_format='%.6f'
    )

    print("\n--- Run Summary ---")
    print(f"Baseline LSTM   F1: {f1_baseline:.5f} | Time: {t_baseline:.2f}s")
    print(f"HHO-LSTM        F1: {f1_hho:.5f} | Time: {t_hho:.2f}s | Improvement: {imp_hho:+.2f}%")
    print(f"MHHO-LSTM       F1: {f1_mhho:.5f} | Time: {t_mhho:.2f}s | Improvement: {imp_mhho:+.2f}%")


def mean_std(x):
    return float(np.mean(x)), float(np.std(x, ddof=1))

mean_b, std_b = mean_std(results_baseline)
mean_h, std_h = mean_std(results_hho)
mean_mh, std_mh = mean_std(results_mhho)

df_summary = pd.DataFrame({
    'Model': ['Baseline LSTM', 'HHO-LSTM', 'MHHO-LSTM'],
    'Mean_F1': [mean_b, mean_h, mean_mh],
    'Std_F1': [std_b, std_h, std_mh],
    'Min_F1': [min(results_baseline), min(results_hho), min(results_mhho)],
    'Max_F1': [max(results_baseline), max(results_hho), max(results_mhho)],
    'Mean_Time_sec': [
        np.mean(execution_times['baseline']),
        np.mean(execution_times['hho']),
        np.mean(execution_times['mhho'])
    ],
    'Std_Time_sec': [
        np.std(execution_times['baseline'], ddof=1),
        np.std(execution_times['hho'], ddof=1),
        np.std(execution_times['mhho'], ddof=1)
    ],
    'N_Runs': [N_RUNS]*3
})

df_summary.to_csv(f'summary_results_{timestamp}.csv', index=False, float_format='%.6f')

curves_hho = np.array(curves_hho)
curves_mhho = np.array(curves_mhho)

df_conv = pd.DataFrame({
    'Iteration': range(1, MAX_ITER + 1),
    'HHO_Mean': np.mean(curves_hho, axis=0),
    'HHO_Std': np.std(curves_hho, axis=0, ddof=1),
    'MHHO_Mean': np.mean(curves_mhho, axis=0),
    'MHHO_Std': np.std(curves_mhho, axis=0, ddof=1)
})

df_conv.to_csv(f'convergence_curves_{timestamp}.csv', index=False, float_format='%.6f')

df_stat = pd.DataFrame({
    'Comparison': ['HHO vs Baseline', 'MHHO vs Baseline', 'MHHO vs HHO'],
    'Mean_Improvement_%': [
        ((mean_h - mean_b) / mean_b) * 100,
        ((mean_mh - mean_b) / mean_b) * 100,
        ((mean_mh - mean_h) / mean_h) * 100
    ],
    'Best_Result': [max(results_hho), max(results_mhho), max(results_mhho)],
    'Worst_Result': [min(results_hho), min(results_mhho), min(results_mhho)]
})

df_stat.to_csv(STAT_FILE, index=False, float_format='%.6f')

print("\n" + "="*70)
print("All results saved safely.")
print("="*70)
```
