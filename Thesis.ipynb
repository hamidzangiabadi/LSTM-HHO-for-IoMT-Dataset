# OUR Decode Function 

```python
import numpy as np

BATCH_CHOICES = [64, 128, 256]

def decode_position(position):

    units = int(np.round(position[0]))
    dropout = float(position[1])
    learning_rate = float(position[2])
    batch_size = int(np.round(position[3]))
    
    units = int(np.clip(units, 16, 256))
    batch_size = int(np.clip(batch_size, 128, 2048))

    dropout = np.clip(dropout, 0.1, 0.5)
    learning_rate = np.clip(learning_rate, 1e-5, 1e-2)

    return {
        "units": units,
        "dropout": dropout,
        "learning_rate": learning_rate,
        "batch_size": batch_size
    }

```

# Creating the model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, LSTM, Dropout, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping

```

```python

def build_lstm_model(input_shape, units, dropout, learning_rate):
    model = Sequential()

    model.add(Input(shape=input_shape))

    model.add(
        LSTM(
            units=units,
            return_sequences=False,
            activation="tanh"
        )
    )

    model.add(Dropout(dropout))

    model.add(Dense(1, activation="sigmoid"))

    optimizer = Adam(learning_rate=learning_rate)

    model.compile(
        optimizer=optimizer,
        loss="binary_crossentropy",
        metrics=[]  
    )

    return model
```

# Fitness Function 

```python
from sklearn.metrics import f1_score, recall_score, precision_score
import numpy as np
import tensorflow as tf
import gc
```

```python
def fitness_function(position):

    tf.random.set_seed(42)
    np.random.seed(42)

    params = decode_position(position)

    model = build_lstm_model(
        input_shape=(1, 45),
        units=params["units"],
        dropout=params["dropout"],
        learning_rate=params["learning_rate"]
    )

    es = tf.keras.callbacks.EarlyStopping(
        monitor="val_loss",
        patience=3,
        restore_best_weights=True,
        verbose=0
    )

    model.fit(
        X_train_sub,
        y_train_sub,
        validation_data=(X_val_sub, y_val_sub),
        epochs=8,                    
        batch_size=params["batch_size"],
        callbacks=[es],
        verbose=0
    )

    y_prob = model.predict(X_val_sub, verbose=0).ravel()

    thresholds = [0.45, 0.50, 0.55]
    f1_list, rec_list, pre_list = [], [], []

    for th in thresholds:
        y_pred = (y_prob > th).astype(int)
        f1_list.append(f1_score(y_val_sub, y_pred))
        rec_list.append(recall_score(y_val_sub, y_pred))
        pre_list.append(precision_score(y_val_sub, y_pred))

    f1_mean = np.mean(f1_list)
    rec_mean = np.mean(rec_list)
    pre_mean = np.mean(pre_list)
    f1_std  = np.std(f1_list)

    fitness = (
        0.60 * f1_mean +
        0.15 * rec_mean +
        0.15 * pre_mean +
        0.10 * (1 - f1_std)
    )

    tf.keras.backend.clear_session()
    del model
    gc.collect()

    return fitness
```

# Levy Flight

```python
import numpy as np
from scipy.special import gamma

def levy_flight(dim, beta=1.5):
    sigma_u = (
        gamma(1 + beta) * np.sin(np.pi * beta / 2) /
        (gamma((1 + beta) / 2) * beta * 2 ** ((beta - 1) / 2))
    ) ** (1 / beta)

    u = np.random.normal(0, sigma_u, dim)
    v = np.random.normal(0, 1, dim)

    step = u / (np.abs(v) ** (1 / beta))
    return step
```

# MHHO

```python
class MHHO:
    def __init__(self, fitness_function, lb, ub, dim, n_hawks=10, max_iter=10,
                 w_max=0.95, w_min=0.4, levy_alpha=0.15, record_curve=True):
        self.fitness_function = fitness_function
        self.lb = lb
        self.ub = ub
        self.dim = dim
        self.n_hawks = n_hawks
        self.max_iter = max_iter
        self.w_max = w_max
        self.w_min = w_min
        self.lev_alpha = levy_alpha
        self.record_curve = record_curve

        self.hawks = np.random.uniform(lb, ub, (n_hawks, dim))
        self.fitness = np.array([fitness_function(h) for h in self.hawks])

        idx = np.argmax(self.fitness)
        self.rabbit_pos = self.hawks[idx].copy()
        self.rabbit_fit = self.fitness[idx]

    def optimize(self):
        best_curve = [] if self.record_curve else None

        for t in range(self.max_iter):
            E0 = 2 * np.random.rand() - 1
            E = 2 * E0 * (1 - t / self.max_iter)
            
            omega = self.w_max - (self.w_max - self.w_min) * (t / self.max_iter)

            for i in range(self.n_hawks):
                J = 2 * (1 - np.random.rand())
                X = self.hawks[i]

                if abs(E) >= 1:
                    rand_idx = np.random.randint(self.n_hawks)
                    X_rand = self.hawks[rand_idx]
                    X_hho = X_rand - np.random.rand() * np.abs(X_rand - 2 * np.random.rand() * X)

                else:
                    X_hho = self.rabbit_pos - E * np.abs(J * self.rabbit_pos - X)
                    
                    if 0.3 < abs(E) < 0.8 and np.random.rand() < 0.5:
                        levy = levy_flight(self.dim) 
                        X_hho = X_hho + self.levy_alpha * levy

                X_new = omega * X + (1 - omega) * X_hho
                X_new = np.clip(X_new, self.lb, self.ub)

                fit_new = self.fitness_function(X_new)

                if fit_new > self.fitness[i]:
                    self.hawks[i] = X_new
                    self.fitness[i] = fit_new

            idx = np.argmax(self.fitness)
            if self.fitness[idx] > self.rabbit_fit:
                self.rabbit_fit = self.fitness[idx]
                self.rabbit_pos = self.hawks[idx].copy()

            if self.record_curve:
                best_curve.append(self.rabbit_fit)

            print(f"Iter {t+1}/{self.max_iter} | Best F1: {self.rabbit_fit:.5f}")

        curve_to_return = np.array(best_curve) if self.record_curve else np.array([])
        
        return self.rabbit_pos, self.rabbit_fit, curve_to_return
```

```python
def evaluate_mhho(
    fitness_function,
    lb,
    ub,
    dim,
    n_runs=5,
    n_hawks=10,
    max_iter=10
):
    best_scores = []

    for run in range(n_runs):
        print(f"\n========== Run {run+1}/{n_runs} ==========")

        mhho = MHHO(
            fitness_function=fitness_function,
            lb=lb,
            ub=ub,
            dim=dim,
            n_hawks=n_hawks,
            max_iter=max_iter
        )

        best_pos, best_fit = mhho.optimize()
        best_scores.append(best_fit)

        print(f"Run {run+1} Best F1: {best_fit:.5f}")

    best_scores = np.array(best_scores)

    print("\n========== MHHO Final Report ==========")
    print(f"Best  F1 : {best_scores.max():.5f}")
    print(f"Mean  F1 : {best_scores.mean():.5f}")
    print(f"Std   F1 : {best_scores.std():.5f}")

    return best_scores
```

```python
def run_baseline_lstm():

    model = build_lstm_model(
        input_shape=(1, 45),
        units=128,
        dropout=0.3,
        learning_rate=1e-3
    )

    model.fit(
        X_train_sub,
        y_train_sub,
        validation_data=(X_val_sub, y_val_sub),
        epochs=EPOCHS_FITNESS,
        batch_size=256,
        verbose=0
    )

    y_pred_prob = model.predict(X_val_sub, verbose=0)
    y_pred = (y_pred_prob > 0.5).astype(int)

    f1 = f1_score(y_val_sub, y_pred)

    tf.keras.backend.clear_session()
    del model
    gc.collect()

    return f1
```

# pure HHO

```python
class HHO:
    def __init__(self, fitness_function, lb, ub, dim,
                 n_hawks=10, max_iter=10, record_curve=False):

        self.fitness_function = fitness_function
        self.lb = lb
        self.ub = ub
        self.dim = dim
        self.n_hawks = n_hawks
        self.max_iter = max_iter
        self.record_curve = record_curve

        self.hawks = np.random.uniform(lb, ub, (n_hawks, dim))
        self.fitness = np.array([fitness_function(h) for h in self.hawks])

        idx = np.argmax(self.fitness)
        self.rabbit_pos = self.hawks[idx].copy()
        self.rabbit_fit = self.fitness[idx]

    def optimize(self):
        best_curve = []

        for t in range(self.max_iter):
            E0 = 2 * np.random.rand() - 1
            E = 2 * E0 * (1 - t / self.max_iter)

            for i in range(self.n_hawks):
                X = self.hawks[i]
                J = 2 * (1 - np.random.rand())

                if abs(E) >= 1:
                    rand_idx = np.random.randint(self.n_hawks)
                    X_rand = self.hawks[rand_idx]
                    X_new = X_rand - np.random.rand() * np.abs(X_rand - 2 * np.random.rand() * X)
                else:
                    X_new = self.rabbit_pos - E * np.abs(J * self.rabbit_pos - X)

                X_new = np.clip(X_new, self.lb, self.ub)
                fit_new = self.fitness_function(X_new)

                if fit_new > self.fitness[i]:
                    self.hawks[i] = X_new
                    self.fitness[i] = fit_new

            idx = np.argmax(self.fitness)
            if self.fitness[idx] > self.rabbit_fit:
                self.rabbit_fit = self.fitness[idx]
                self.rabbit_pos = self.hawks[idx].copy()

            best_curve.append(self.rabbit_fit)

            print(f"[HHO] Iter {t+1}/{self.max_iter} | Best F1: {self.rabbit_fit:.5f}")

        return self.rabbit_pos, self.rabbit_fit, np.array(best_curve)
```

# Run Tests

```python
import numpy as np
import random
import tensorflow as tf
import pandas as pd
from datetime import datetime
import time

def set_seed(seed):
    np.random.seed(seed)
    random.seed(seed)
    tf.random.set_seed(seed)

N_RUNS = 5
dim = 4

curves_hho = []
curves_mhho = []

results_baseline = []
results_hho = []
results_mhho = []

detailed_results = []
execution_times = {'baseline': [], 'hho': [], 'mhho': []}

N_HAWKS = 10     
MAX_ITER = 15

for run in range(N_RUNS):
    seed = 100 + run
    set_seed(seed)
    print(f"\n================ RUN {run+1}/{N_RUNS} | Seed = {seed} ================")

    print("Running Baseline LSTM ...")
    start_time = time.time()
    f1_baseline = run_baseline_lstm() 
    baseline_time = time.time() - start_time
    results_baseline.append(f1_baseline)
    execution_times['baseline'].append(baseline_time)

    print("\nRunning HHO-LSTM ...")
    start_time = time.time()
    hho = HHO(
        fitness_function=fitness_function, 
        lb=lb,
        ub=ub,
        dim=dim,
        n_hawks=N_HAWKS,
        max_iter=MAX_ITER,
        record_curve=True  
    )
    best_pos_hho, best_f1_hho, best_curve_hho = hho.optimize()
    hho_time = time.time() - start_time
    results_hho.append(best_f1_hho)
    curves_hho.append(best_curve_hho)
    execution_times['hho'].append(hho_time)

    print("\nRunning MHHO-LSTM ...")
    start_time = time.time()
    mhho = MHHO(
        fitness_function=fitness_function, 
        lb=lb,
        ub=ub,
        dim=dim,
        n_hawks=N_HAWKS,
        max_iter=MAX_ITER,
        record_curve=True
    )
    best_pos_mhho, best_f1_mhho, best_curve_mhho = mhho.optimize()
    mhho_time = time.time() - start_time
    results_mhho.append(best_f1_mhho)
    curves_mhho.append(best_curve_mhho)
    execution_times['mhho'].append(mhho_time)

    improvement_hho = ((best_f1_hho - f1_baseline) / f1_baseline) * 100
    improvement_mhho = ((best_f1_mhho - f1_baseline) / f1_baseline) * 100

    detailed_results.append({
        'Run': run + 1,
        'Seed': seed,
        'Baseline_F1': f1_baseline,
        'HHO_F1': best_f1_hho,
        'MHHO_F1': best_f1_mhho,
        'Baseline_Time_sec': baseline_time,
        'HHO_Time_sec': hho_time,
        'MHHO_Time_sec': mhho_time,
        'HHO_Improvement_%': improvement_hho,
        'MHHO_Improvement_%': improvement_mhho
    })

    print("\n--- Run Summary ---")
    print(f"Baseline LSTM   F1: {f1_baseline:.5f} | Time: {baseline_time:.2f}s")
    print(f"HHO-LSTM        F1: {best_f1_hho:.5f} | Time: {hho_time:.2f}s | Improvement: {improvement_hho:+.2f}%")
    print(f"MHHO-LSTM       F1: {best_f1_mhho:.5f} | Time: {mhho_time:.2f}s | Improvement: {improvement_mhho:+.2f}%")

def mean_std(arr):
    return float(np.mean(arr)), float(np.std(arr, ddof=1))  

mean_b, std_b = mean_std(results_baseline)
mean_h, std_h = mean_std(results_hho)
mean_mh, std_mh = mean_std(results_mhho)

mean_time_b, std_time_b = mean_std(execution_times['baseline'])
mean_time_h, std_time_h = mean_std(execution_times['hho'])
mean_time_mh, std_time_mh = mean_std(execution_times['mhho'])

timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")

df_detailed = pd.DataFrame(detailed_results)
df_detailed.to_csv(f'detailed_results_{timestamp}.csv', index=False, float_format='%.6f')
print(f"\nâœ“ Detailed results saved to: detailed_results_{timestamp}.csv")

summary_data = {
    'Model': ['Baseline LSTM', 'HHO-LSTM', 'MHHO-LSTM'],
    'Mean_F1': [mean_b, mean_h, mean_mh],
    'Std_F1': [std_b, std_h, std_mh],
    'Min_F1': [min(results_baseline), min(results_hho), min(results_mhho)],
    'Max_F1': [max(results_baseline), max(results_hho), max(results_mhho)],
    'Mean_Time_sec': [mean_time_b, mean_time_h, mean_time_mh],
    'Std_Time_sec': [std_time_b, std_time_h, std_time_mh],
    'N_Runs': [N_RUNS, N_RUNS, N_RUNS]
}
df_summary = pd.DataFrame(summary_data)
df_summary.to_csv(f'summary_results_{timestamp}.csv', index=False, float_format='%.6f')
print(f"âœ“ Summary results saved to: summary_results_{timestamp}.csv")

np.save('curve_hho.npy', np.array(curves_hho, dtype=float))
np.save('curve_mhho.npy', np.array(curves_mhho, dtype=float))
print("âœ“ Convergence curves saved to: curve_hho.npy & curve_mhho.npy")

curves_hho_array = np.array(curves_hho)
curves_mhho_array = np.array(curves_mhho)

mean_curve_hho = np.mean(curves_hho_array, axis=0)
std_curve_hho = np.std(curves_hho_array, axis=0, ddof=1)
mean_curve_mhho = np.mean(curves_mhho_array, axis=0)
std_curve_mhho = np.std(curves_mhho_array, axis=0, ddof=1)

df_convergence = pd.DataFrame({
    'Iteration': range(1, MAX_ITER + 1),
    'HHO_Mean': mean_curve_hho,
    'HHO_Std': std_curve_hho,
    'MHHO_Mean': mean_curve_mhho,
    'MHHO_Std': std_curve_mhho
})
df_convergence.to_csv(f'convergence_curves_{timestamp}.csv', index=False, float_format='%.6f')
print(f"convergence_curves_{timestamp}.csv")

statistical_comparison = pd.DataFrame({
    'Comparison': ['HHO vs Baseline', 'MHHO vs Baseline', 'MHHO vs HHO'],
    'Mean_Improvement_%': [
        ((mean_h - mean_b) / mean_b) * 100,
        ((mean_mh - mean_b) / mean_b) * 100,
        ((mean_mh - mean_h) / mean_h) * 100
    ],
    'Best_Result': [
        max(results_hho),
        max(results_mhho),
        max(results_mhho)
    ],
    'Worst_Result': [
        min(results_hho),
        min(results_mhho),
        min(results_mhho)
    ]
})
statistical_comparison.to_csv(f'statistical_comparison_{timestamp}.csv', index=False, float_format='%.6f')
print(f"statistical_comparison_{timestamp}.csv")

print("\n" + "="*70)

```
